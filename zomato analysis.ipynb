```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

PRINT ENTIRE DATA SET

```python
df = pd.read_csv(r"c:\Users\LENOVO\Downloads\Zomato-data-.csv")
print(df.to_string())
```

converting rate in only float value

```python
def handleRate(value):
    value = str(value).split('/')
    
    return float(value[0])

df['rate']=df['rate'].apply(handleRate)
print(df.to_string())

```

information of the data set

```python
print(df.info())
```

checking type of restaurants used most

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("Types of restaurant")
plt.ylabel("Number of restaurants")
plt.show
```

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result,marker="o")
plt.xlabel("types of restaurants")
plt.ylabel("votes")

```

```python
max_votes = df['votes'].max()
restaurant_with_max_votes = df.loc[df['votes'] == max_votes, 'name']
print("restaurant with the maximum votes :")
print(restaurant_with_max_votes )
```

```python
sns.countplot(x=df['online_order'])
plt.xlabel("online_order")

plt.show()
```

```python
plt.hist(df['rate'],bins=5)
plt.title('ratings')
plt.show()
```

```python
sns.countplot(x=df['approx_cost(for two people)'])
plt.show()
```

```python
sns.boxplot(x='online_order', y='rate', data=df)
plt.show()
```

```python
table = df.table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(table, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('Listed In (Type)')
plt.show()

```
